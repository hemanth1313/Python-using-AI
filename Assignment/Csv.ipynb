```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
filename = 'IMDB.csv'
x=[]
y=[]

with open ('IMDB.csv','r') as csvFile:
    plots = csv.DictReader(csvFile)
    for row in plots:
        if row["Year"]== '2013':
            x.append(row['Year'])
            y.append(row['Rating']) 
plt.scatter(x,y)
plt.show()

        
```
